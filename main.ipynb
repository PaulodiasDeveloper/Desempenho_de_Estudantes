```python
# importanção das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importar e entendendo base de dados

```python
# Lendo arquivo CSV
df = pd.read_csv('student_habits_performance.csv')
```

```python
# Visualizando dados
df
```

## Quais hábitos impactam mais o desempenho das colunas?

```python
# Tipo de dados
df.info()
```

```python
# Filtros das coluna númericas para análise
cols = [
    "study_hours_per_day",
    "social_media_hours",
    "netflix_hours",
    "sleep_hours",
    "attendance_percentage",
    "exercise_frequency",
    "mental_health_rating",
    "exam_score"
]
# Plotar mapa de calor (heatmap)
sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlação entre hábitos e nota final')
plt.show()
```

## Alunos que mais estudam têm melhor desempenho?

```python
# Grtáfico de dispersão com linha de regressão
# x="study_hours_per_day", y="exam_score"
sns.lmplot(data=df, x="study_hours_per_day", y="exam_score")
plt.title("Mais estudo → notas mais altas?")
plt.show()
```

```python
# Correlação entre horas de estudo e nota final
# Comparando médias: quem esutuda >5h x <2h
filtro_estudo_alto = df["study_hours_per_day"] > 5
filtro_estudo_baixo = df["study_hours_per_day"] < 2

grupo_estudo_alto = df[filtro_estudo_alto]["exam_score"]
grupo_estudo_baixo = df[filtro_estudo_baixo]["exam_score"]

print("Média de notas dos que estudam mais de 5h/dia:", grupo_estudo_alto.mean())
print("Média de notas dos que estudam menos de 2h/dia:", grupo_estudo_baixo.mean())
```

## O tempo gasto de redes sociais afeta o desempenho dos alunos?

```python
# Redes sociais:  distribuição geral (histograma)
# x="social_media_hours"
sns.histplot(data=df, x="social_media_hours", bins=10, kde=True)
plt.title("Distribuição do uso de redes sociais")
plt.show()
```

```python
# Avaliando notas médias
# por diferentes intervalos (bins) de períodos gastos em redes sociais
# ["0-2h", "2h-4h", "4h-6h", "6h+"]
df["social_media_bin"] = pd.cut(
    df["social_media_hours"],
    bins=[0, 2, 4, 6],
    labels=["0-2h", "2h-4h", "4h-6h"]
    )

# Plotar gráfico de caixa (boxplot)
sns.boxplot(x="social_media_bin", y="exam_score", data=df)
plt.title("Notas por tempo gasto em redes sociais")
plt.show()
```

## Alunos mais saudáveis têm melhores desempenho?

```python
# Frequência de exercícios físicos
for col in ["exercise_frequency", "mental_health_rating", "diet_quality"]:
    sns.boxplot(x=col, y="exam_score", data=df)
    plt.title(f"Notas por {col.replace('_', ' ')}")
    plt.show()
```

## Há diferença na notas entre mulheres e homens?

```python
# Estatítisca por gênero (media e desvio padrão)
df.groupby(["gender"])["exam_score"].agg(['mean', 'std'])
```
